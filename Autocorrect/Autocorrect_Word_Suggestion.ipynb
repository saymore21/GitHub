```python
from collections import Counter
import numpy as np
import nltk
import re

source = ""
target = ""
sourcelen = 0
targetlen = 0


file = open('big.txt').read()
plain_text = re.sub(r'[^\w\s]','',file).lower()

words = nltk.word_tokenize(plain_text)
Words = set(words)

big_bigrams = list(nltk.bigrams(nltk.word_tokenize(plain_text)))
count_times = Counter(words)

word_len = {}
for word in Words:
    word_len[word] = len(word)
```

```python
def find_alt(word):
    
    #finding length of the word
    size = len(word)
    
    word_list = []    
    similar_len_word_list = []
    similar_word_list = []
    for key, length in word_len.items():
        if(length == size):
            similar_len_word_list.append(key)           
        
    for x in similar_len_word_list:
        similar = 0
        for i in word:
            if(i in x):
                similar += 1
        if(similar == len(word)):
            similar_word_list.append(x)
    
    for key, length in word_len.items():
        if(length == size or length == size + 1 or length == size - 1):
            word_list.append(key)
            
    edit_dist_dict = {}
    for words in word_list:
        edit_dist_dict[words] = find_min_edits(word,words)
        
    min_edit_list = []
    min_edit_freq_list = {}
    
    for key, edit in edit_dist_dict.items():
        if(edit == 0  or edit == 1 or edit == 2):
            min_edit_list.append(key)
            
    for key in min_edit_list:
        for keys, count in count_times.items():
            if(keys == key):
                min_edit_freq_list[key] = count
            

    min_edit_freq_list = sorted(min_edit_freq_list, key=min_edit_freq_list.get, reverse=True)
    
    top_similar_list  = []
    count = 0
    for item in min_edit_freq_list:
        top_similar_list.append(item)
        count += 1
        if(count  == 10):
            break
            
    top_similar_edit_dict = {}
    edit_final_perfectlist = {}
    
    for n in similar_word_list:
        for key, edit in edit_dist_dict.items():
            if(n == key):
                edit_final_perfectlist[key] = edit
    
    similar_word_list = sorted(edit_final_perfectlist, key = edit_final_perfectlist.get, reverse=False)
    
    for elem in similar_word_list:
        if(elem in top_similar_list):
            top_similar_list.remove(elem)
    
    for n in top_similar_list:
        for key, edit in edit_dist_dict.items():
            if(n == key):
                top_similar_edit_dict[key] = edit
                
    top_similar_list = sorted(top_similar_edit_dict, key = top_similar_edit_dict.get, reverse=False)
    print(similar_word_list + top_similar_list)

        
        
    

def find_min_edits(sourcestring, targetstring):
    
    global source, target, sourcelen, targetlen
    #Finding the length of source and target strings
    sourcelen = len(sourcestring)
    targetlen = len(targetstring)
    
    source = sourcestring
    target = targetstring
    #Initializing a matrix to calculate minimum edit dstance
    minmatrix = np.zeros(((targetlen + 1),(sourcelen + 1)))
    
    #code to intialize the first row and column which shows the minimum edits required 
    #to get the source and target strings if eeither of them was an empty string
    rowinit = 0
    columninit = 0
    for column in range(sourcelen + 1):
        minmatrix[0][column] = rowinit
        rowinit += 1
        
    for row in range(targetlen + 1):
        minmatrix[row][0] = columninit
        columninit += 1
        
    #code to calculate the rest of marix for getting target string when source string is given
    for row in range(1, targetlen + 1):
        for column in range(1, sourcelen + 1):
            if(target[row - 1] == source[column - 1]):
                minmatrix[row][column] = minmatrix[row - 1][column - 1]
            elif (target[row - 1] != source[column - 1]):
                minmatrix[row][column] = (min(minmatrix[row - 1][column - 1], minmatrix[row][column - 1],
                                              minmatrix[row - 1][column]) + 1)
                
    return int(minmatrix[targetlen][sourcelen])

#Main Function
if __name__ == '__main__':
    word = input("Enter the word you need suggestions for: ")
    find_alt(word)
```
